```python
# loading the libraries

from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.linear_model import LogisticRegression

from sklearn.model_selection import train_test_split
import pandas as pd
from sklearn.metrics import confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python

# Read the Iris dataset from CSV
iris_df = pd.read_csv('iris.csv')

# Separate features (X) and target variable (y)
X = iris_df.drop(columns=['species'])  # Assuming 'species' column contains the target variable
y = iris_df['species']

# Split the data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
```

```python
# Create a pipeline with three steps: scaling, PCA, and logistic regression
pipe = Pipeline([
    ('scaler', StandardScaler()),
    ('pca', PCA(n_components=2)),
    ('classifier', LogisticRegression())
])

# Fit the pipeline to the training data
pipe.fit(X_train, y_train)
```

```python
# Evaluate the pipeline on the test data
score = pipe.score(X_test, y_test)
print(f'Test accuracy: {score:.3f}')
```

```python
#!pip install seaborn
```

```python

# Predictions on the test set
predictions = pipe.predict(X_test)

# Compute confusion matrix
cm = confusion_matrix(y_test, predictions)

# Plot confusion matrix
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=np.unique(y), yticklabels=np.unique(y))
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix (Logistic Regression')
plt.show()
```

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.ensemble import RandomForestClassifier  # New import
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

# Read the Iris dataset from CSV
iris_df = pd.read_csv('iris.csv')

# Separate features (X) and target variable (y)
X = iris_df.drop(columns=['species'])  # Assuming 'species' column contains the target variable
y = iris_df['species']

# Split the data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)

# Create a pipeline with three steps: scaling, PCA, and RandomForestClassifier
pipe_rf = Pipeline([
    ('scaler', StandardScaler()),
    ('pca', PCA(n_components=2)),
    ('classifier', RandomForestClassifier())  # Using RandomForestClassifier instead of LogisticRegression
])

# Fit the pipeline to the training data
pipe_rf.fit(X_train, y_train)



```

```python
# Evaluate the pipeline on the test data
y_pred_rf = pipe_rf.predict(X_test)
score_rf = accuracy_score(y_test, y_pred_rf)
print(f'Test accuracy (Random Forest): {score_rf:.3f}')
```

```python
# Importing necessary libraries
from sklearn.metrics import confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

# Compute confusion matrix for Random Forest Classifier
cm_rf = confusion_matrix(y_test, y_pred_rf)

# Plot confusion matrix for Random Forest Classifier
plt.figure(figsize=(8, 6))
sns.heatmap(cm_rf, annot=True, fmt='d', cmap='Blues', xticklabels=np.unique(y), yticklabels=np.unique(y))
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix (Random Forest)')
plt.show()
```

```python
import matplotlib.pyplot as plt

# Test accuracy scores for Logistic Regression and Random Forest
test_accuracy_scores = [score, score_rf]

# Labels for the bar plot
models = ['Logistic Regression', 'Random Forest']

# Define a custom color scheme
colors = ['#1f77b4', '#ff7f0e']  # Blue and orange

# Create bar plot with custom colors
plt.figure(figsize=(8, 6))
plt.bar(models, test_accuracy_scores, color=colors)
plt.xlabel('Model')
plt.ylabel('Test Accuracy')
plt.title('Comparison of Test Accuracy')
plt.ylim(0, 1)  # Set y-axis limit for better visualization
plt.show()
```
